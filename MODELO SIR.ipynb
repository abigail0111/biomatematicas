```python
#AQUI VAMOS HACER LA SIMULACION DE LA SOLUCION DEL MODELOS SIR 

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

```

```python
#parametros del modelo

beta,gamma = 0.4,0.7
tmax= 365
t = np.linspace(0,tmax,tmax*10)  #el ultimo es el numero de puntos 
```

```python
#definicion del lado derecho de la ecuacion diferencial 

def modSIR(y,t,beta,gamma):
    S, I, R = y
    dSdt = -beta*S*I
    dIdt = beta*S*I - gamma*I
    dRdt = gamma*I
    return dSdt,dIdt,dRdt
```

```python
#condiciones inciales 

S0,I0,R0 = 25,15,0
Y0 = S0,I0,R0

solucion = odeint(modSIR,Y0,t,args=(beta,gamma))
S,I,R=solucion.T
```

```python
#grafica de la solucion

%matplotlib inline
fig = plt.figure()
plt.plot(t,S,'b', label="suceptibles")
plt.plot(t,I,'r', label="Infecciosos")
plt.plot(t,R,'g', label="Recuperados")
plt.xlabel("tiempo")
plt.ylabel("numero")
#plt.ylim(0,1,2)
plt.xlim(0,10)
legend =fig.legend()
legend.get_frame().set_alpha(0.5)
```
